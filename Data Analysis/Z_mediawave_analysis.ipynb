## Interpretasi definisi kolom

**Open**: Harga pembukaan saham pada hari perdagangan tertentu. Ini adalah harga pertama yang ditetapkan saat perdagangan dimulai pada hari itu.

**High**: Harga tertinggi yang dicapai oleh saham selama periode perdagangan tertentu. Ini mencerminkan harga tertinggi yang dibayar oleh pembeli selama hari itu.

**Low**: Harga terendah yang dicapai oleh saham selama periode perdagangan tertentu. Ini mencerminkan harga terendah yang diterima oleh penjual selama hari itu.

**Close**: Harga penutupan saham pada hari perdagangan tertentu. Ini adalah harga terakhir yang ditetapkan sebelum pasar ditutup.

**Adjusted Close**: Harga penutupan saham yang disesuaikan untuk memperhitungkan perubahan seperti dividen, pembagian saham, atau perubahan struktur modal. Ini memberikan gambaran yang lebih akurat tentang kinerja saham dari waktu ke waktu.

**Volume**: Jumlah total saham yang diperdagangkan selama periode perdagangan tertentu. Ini mencerminkan seberapa aktifnya perdagangan saham pada hari itu.

Jika kita memiliki data tambahan seperti harga pembelian dan jumlah saham yang dibeli untuk setiap transaksi, kita dapat menghitung total_profit dan total_invested dengan langkah-langkah berikut:

Hitung nilai investasi untuk setiap transaksi:

1. Nilai Investasi = Harga Pembelian * Jumlah Saham yang Dibeli
Hitung total investasi dengan menjumlahkan nilai investasi untuk semua transaksi.

1. Hitung total nilai penjualan untuk setiap transaksi:

1. Nilai Penjualan = Harga Penjualan * Jumlah Saham yang Dijual
Hitung total penjualan dengan menjumlahkan nilai penjualan untuk semua transaksi.

1. Hitung total_profit dengan mengurangkan total investasi dari total penjualan:

1. Total Profit = Total Penjualan - Total Investasi

```python
# Cek Data Outlier
from sklearn.ensemble import IsolationForest

# Menghilangkan kolom yang tidak ingin digunakan untuk deteksi outlier
features = df.drop(columns=['Date', 'Company', 'Sector'])  # Misalnya, menghilangkan kolom 'Date', 'Company', dan 'Sector'

# Membuat model Isolation Forest
model = IsolationForest(contamination=0.05)  # Persentase outlier yang diharapkan

# Melatih model pada data
model.fit(features)

# Memprediksi outlier pada data
outlier_labels = model.predict(features)

# Menambahkan label outlier ke dalam DataFrame
df['Outlier'] = outlier_labels

# Menampilkan data yang merupakan outlier
outliers = df[df['Outlier'] == -1]
outliers
```

Deteksi outlier untuk keperluan analisis menggunakan metode Isolation Forest bekerja dengan prinsip bahwa outlier cenderung menjadi "terpencil" atau "terisolasi" dari mayoritas data. Metode ini membangun sejumlah decision secara acak, dan pada setiap tahap, memilih secara acak satu fitur dan satu nilai ambang acak antara nilai minimum dan maksimum dari fitur tersebut. Proses ini terus berlanjut sampai setiap data ditempatkan dalam satu daun atau sampai memenuhi kriteria penghentian. Kemudian, skor anomali dihitung berdasarkan rata-rata panjang jalur yang dibutuhkan oleh data untuk mencapai daun tertentu di semua pohon.

Penghapusan kolom 'Date', 'Company', dan 'Sector' dilakukan karena:

1. Kolom 'Date': Biasanya tidak relevan untuk deteksi outlier. Outlier dalam konteks harga saham mungkin berkaitan dengan pergerakan harga atau volatilitas pasar, bukan tanggal tertentu.

2. Kolom 'Company': Jika digunakan sebagai fitur dalam deteksi outlier, akan sulit bagi model untuk membedakan antara outlier dari satu perusahaan dengan outlier dari perusahaan lain. Selain itu, variabilitas harga saham dapat bervariasi secara signifikan antara perusahaan.

3. Kolom 'Sector': Sama seperti kolom 'Company', penggunaan informasi ini mungkin tidak relevan atau membuat deteksi outlier lebih rumit. Outlier dapat terjadi di setiap sektor industri dan tidak selalu berkaitan dengan sektor tertentu.

Dengan menghapus kolom-kolom ini, fokus deteksi outlier berada pada fitur-fitur yang lebih relevan, seperti harga saham itu sendiri. Ini memungkinkan model untuk lebih fokus pada pola-pola yang mungkin menandakan adanya outlier dalam data harga saham.

Dalam konteks pembentukan portofolio saham, penting untuk melakukan manajemen data outlier dengan cermat. Langkah-langkah seperti missing values, duplicate data, data invalid dan penanganan outlier perlu dilakukan untuk memastikan bahwa portofolio saham yang dibentuk didasarkan pada data yang bersih, akurat, dan representatif. Hal ini akan membantu investor membuat keputusan investasi yang lebih baik dan membangun portofolio yang sesuai dengan tujuan investasi dan toleransi risiko mereka.

```python
def plot_var_vs_target(df, var_analyze, var_target, show_num=20, size_h=14, size_e=6):

    df_plot = (df.groupby([var_analyze, var_target])
                              .size().reset_index().pivot(columns=var_target, index=var_analyze, values=0))
    df_plot = df_plot.fillna(0)
    df_plot.plot(kind='bar',  stacked=True, figsize=(size_h,size_e));


def plot_var_vs_target_normalized(df, var_analyze, var_target, show_num=20, size_h=14, size_e=6):

    df_plot = (df.groupby([var_analyze, var_target])
                              .size().reset_index().pivot(columns=var_target, index=var_analyze, values=0))

    df_plot = df_plot.fillna(0)

    cols = df_plot.columns.tolist()
    df_plot['total'] = 0
    for col in cols:
        df_plot['total'] = df_plot['total']+ df_plot[col]

    cols = df_plot.columns.tolist()[:-1]

    for col in cols:
        df_plot[col] = df_plot[col] / df_plot['total']

    df_plot = df_plot.drop(columns=['total'])

    df_plot.plot(kind='bar', stacked=True, figsize=(size_h,size_e));
```
